```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class MatrixGenerator(widgets.VBox):
    def __init__(self, rows=3, cols=3):
        self.rows = widgets.IntSlider(value=rows, min=1, max=10, description="Filas")
        self.cols = widgets.IntSlider(value=cols, min=1, max=10, description="Columnas")
        self.generate_btn = widgets.Button(description="Generar matriz")
        self.output = widgets.Output()
        self.grid = None

        self.generate_btn.on_click(self._generate_grid)
        super().__init__([self.rows, self.cols, self.generate_btn, self.output])
        self._generate_grid(None)

    def _generate_grid(self, b):
        with self.output:
            clear_output()
            self.grid = []
            for i in range(self.rows.value):
                row = []
                for j in range(self.cols.value):
                    cell = widgets.BoundedIntText(value=0, min=0, max=99, layout=widgets.Layout(width="40px"))
                    row.append(cell)
                self.grid.append(row)
            grid_box = widgets.VBox([widgets.HBox(row) for row in self.grid])
            display(grid_box)
            display(widgets.Button(description="Obtener matriz", 
                                   on_click=self._show_matrix))

    def _show_matrix(self):
        matrix = [[cell.value for cell in row] for row in self.grid]
        print("Matriz generada:")
        print(matrix)

# Ejemplo de uso en una celda de Jupyter:
gen = MatrixGenerator()
display(gen)
```
