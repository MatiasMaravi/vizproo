```python
import seaborn as sns
from vizproo import  ScatterPlot, RadViz, StarCoordinates, Button, MatrixLayout
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset iris de seaborn
iris = sns.load_dataset("iris")
iris.head()
```

```python
from vizproo import MatrixCreator
m =MatrixCreator(
    rows=5,
    columns=4
)
m
```

```python
plot1 = RadViz(data=iris, dimensions=['sepal_length','sepal_width','petal_width','petal_length'], hue='species')
plot2 = StarCoordinates(data=iris, dimensions=['sepal_length','sepal_width','petal_width','petal_length'], hue='species')

button = Button(description="Calculate TSNE")

scatterplot = ScatterPlot(data=pd.DataFrame(), x='dim1', y='dim2', hue='color')

def on_plot1_selected(change):
    df = plot1.selectedValues
    # Actualiza plot2 cuando HAY selección; si no, puedes limpiar o restaurar
    if not df.empty:
        plot2.data = df
    else:
        plot2.data = iris  # o pd.DataFrame()

plot1.on_select_values(on_plot1_selected)

def calculate_TSNE(_):
    # Usa la selección si existe; si no, usa todo el iris
    df = plot1.selectedValues if not plot1.selectedValues.empty else iris
    X = df[plot1.dimensions].values       # solo columnas numéricas
    y = df['species'].astype(str).values  # etiquetas alineadas con X

    # Asegura una perplexity válida: < n_samples
    n = len(df)
    if n < 3:
        return  # muy pocos puntos para t-SNE
    perplexity = min(30, max(5, n - 1))

    tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity)
    X_tsne = tsne.fit_transform(X)

    tsne_df = pd.DataFrame({
        'dim1': X_tsne[:, 0],
        'dim2': X_tsne[:, 1],
        'color': y
    })
    scatterplot.data = tsne_df

button.on_click(calculate_TSNE)

matrix = m.data
matrixLayout = MatrixLayout(matrix)
matrixLayout.add(plot1, 1)
matrixLayout.add(plot2, 2)
matrixLayout.add(button, 3)
matrixLayout.add(scatterplot, 4)
matrixLayout
```
